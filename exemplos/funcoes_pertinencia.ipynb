```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

x = np.arange(0, 10, 0.1)
```

# Função de Pertinênica Triangular

```python
def plot_function_tri(x, y1, y2, y3):
    plt.figure(figsize=(5, 3))
    plt.plot(x, y1, label="y1")
    plt.plot(x, y2, label="y2")
    plt.plot(x, y3, label="y3")
    plt.xlabel("X")
    plt.ylabel("Pertinência")
    plt.title("Função de Pertinência Triangular")
    plt.legend()
    plt.show()
```

```python
y1 = fuzz.trimf(x, [3, 4, 5])
y2 = fuzz.trimf(x, [2, 4, 7])
y3 = fuzz.trimf(x, [1, 4, 9])

plot_function_tri(x, y1, y2, y3)
```

```python
y1 = fuzz.trimf(x, [2, 3, 5])
y2 = fuzz.trimf(x, [3, 4, 7])
y3 = fuzz.trimf(x, [4, 5, 9])

plot_function_tri(x, y1, y2, y3)
```

# Função de Pertinênica Trapezoidal

```python
def plot_function_trap(x, y1, y2, y3, y4):
    plt.figure(figsize=(5, 3))
    plt.plot(x, y1, label="y1")
    plt.plot(x, y2, label="y2")
    plt.plot(x, y3, label="y3")
    plt.plot(x, y4, label="y4")
    plt.xlabel("X")
    plt.ylabel("Pertinência")
    plt.title("Função de Pertinência Trapezoidal")
    plt.legend()
    plt.show()
```

```python
y1 = fuzz.trapmf(x, [2, 3, 7, 9])
y2 = fuzz.trapmf(x, [3, 4, 6, 8])
y3 = fuzz.trapmf(x, [4, 5, 5, 7])
y4 = fuzz.trapmf(x, [5, 6, 7, 7])

plot_function_trap(x, y1, y2, y3, y4)
```

# Função de Pertinênica Gaussiana

```python
def plot_function_gauss(x, y1, y2, y3, y4):
    plt.figure(figsize=(5, 3))
    plt.plot(x, y1, label="y1")
    plt.plot(x, y2, label="y2")
    plt.plot(x, y3, label="y3")
    plt.plot(x, y4, label="y4")
    plt.xlabel("X")
    plt.ylabel("Pertinência")
    plt.title("Função de Pertinência Gaussiana")
    plt.legend()
    plt.show()
```

```python
y1 = fuzz.gaussmf(x, 5, 0.5)
y2 = fuzz.gaussmf(x, 5, 1)
y3 = fuzz.gaussmf(x, 5, 2)
y4 = fuzz.gaussmf(x, 5, 3)

plot_function_gauss(x, y1, y2, y3, y4)
```

```python
y1 = fuzz.gaussmf(x, 2, 1)
y2 = fuzz.gaussmf(x, 4, 1)
y3 = fuzz.gaussmf(x, 6, 1)
y4 = fuzz.gaussmf(x, 8, 1)

plot_function_gauss(x, y1, y2, y3, y4)
```
