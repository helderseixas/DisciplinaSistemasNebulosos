```python

import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
import skfuzzy.control as ctrl
from sklearn.metrics import mean_squared_error
```

# Função para se aproximar: y = x²

```python
x = np.arange(-2, 2, 0.01, )
y = x ** 2
```

# Tentativa por funções triangulares

# Antecedent

```python
a = ctrl.Antecedent(x, "A")
a['a1'] = fuzz.trimf(x, [-2,-2,-1.25])
a['a2'] = fuzz.trimf(x, [-1.75,-1,-1])
a['a3'] = fuzz.trimf(x, [-1,-0.75,-0.5])
a['a4'] = fuzz.trimf(x, [-0.75,-0.5,-0.25])
a['a5'] = fuzz.trimf(x, [-0.5,0,0.5])
a['a6'] = fuzz.trimf(x, [0.25,0.5,0.75])
a['a7'] = fuzz.trimf(x, [0.5,0.75,1])
a['a8'] = fuzz.trimf(x, [1,1,1.75])
a['a9'] = fuzz.trimf(x, [1.25,2,2])

a.view()
```

# Consequent

```python
b = ctrl.Consequent(np.arange(y.min(), y.max(), 0.01), "B")
b['b1'] = fuzz.trimf(b.universe, [0,0,0.25])
b['b2'] = fuzz.trimf(b.universe, [0.0625,0.25,0.5625])
b['b3'] = fuzz.trimf(b.universe, [0.25,0.5625,1])
b['b4'] = fuzz.trimf(b.universe, [1,1,3.0625])
b['b5'] = fuzz.trimf(b.universe, [1.5625,4,4])

b.view()
```

```python
r1 = ctrl.Rule(antecedent=a['a1'], consequent=b['b5'], label='R1')
r2 = ctrl.Rule(antecedent=a['a2'], consequent=b['b4'], label='R2')
r3 = ctrl.Rule(antecedent=a['a3'], consequent=b['b3'], label='R3')
r4 = ctrl.Rule(antecedent=a['a4'], consequent=b['b2'], label='R4')
r5 = ctrl.Rule(antecedent=a['a5'], consequent=b['b1'], label='R5')
r6 = ctrl.Rule(antecedent=a['a6'], consequent=b['b2'], label='R6')
r7 = ctrl.Rule(antecedent=a['a7'], consequent=b['b3'], label='R7')
r8 = ctrl.Rule(antecedent=a['a8'], consequent=b['b4'], label='R8')
r9 = ctrl.Rule(antecedent=a['a9'], consequent=b['b5'], label='R9')

control_system = ctrl.ControlSystem(rules = [r1, r2, r3, r4, r5, r6, r7, r8, r9])
simulation = ctrl.ControlSystemSimulation(control_system)

def simulate(simulation, xi):
    simulation.input['A'] = xi
    simulation.compute()
    print('z = ',simulation.output['B'])
    a.view(sim=simulation)
    b.view(sim=simulation)
```

# Simulation to x = 0

```python
simulate(simulation, 0)
```

# Simulation to x = 0.25

```python
simulate(simulation, 0.25)
```

# Simulation to x = 0.5

```python
simulate(simulation, 0.5)
```

# Simulation to x = 0.75

```python
simulate(simulation, 0.75)
```

# Simulation to x = 1

```python
simulate(simulation, 1)
```

# Simulation to x = 1.25

```python
simulate(simulation, 1.25)
```

# Simulation to x = 1.5

```python
simulate(simulation, 1.5)
```

# Simulation to x = 1.75

```python
simulate(simulation, 1.75)
```

# Simulation to x = 2

```python
simulate(simulation, 2)
```

# Surface chart

```python
result_simulation = []

for i in x:
    simulation.input['A'] = i
    simulation.compute()
    result_simulation.append(simulation.output['B'])

plt.figure(figsize=(5, 3))
plt.plot(x, result_simulation, label='Simulation')
plt.plot(x, y, label='x²')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

# Mean squared error (MSE)

```python
mse = mean_squared_error(y, result_simulation)
print('MSE: ', mse)
```

# Tentativa por funções gaussianas

# Antecedent

```python
a = ctrl.Antecedent(x, "A")
a['a1'] = fuzz.gaussmf(x, -2, 0.375)
a['a2'] = fuzz.gaussmf(x, -1.1875, 0.0938)
a['a3'] = fuzz.gaussmf(x, -0.75, 0.125)
a['a4'] = fuzz.gaussmf(x, -0.5, 0.125)
a['a5'] = fuzz.gaussmf(x, 0, 0.25)
a['a6'] = fuzz.gaussmf(x, 0.5, 0.125)
a['a7'] = fuzz.gaussmf(x, 0.75, 0.125)
a['a8'] = fuzz.gaussmf(x, 1.1875, 0.0938)
a['a9'] = fuzz.gaussmf(x, 2, 0.375)

a.view()
```

# Consequent

## Tipos de operadores de desfuzzificação:
* ‘centroid’: Centroid of area
* ‘bisector’: bisector of area
* ‘mom’ : mean of maximum
* ‘som’ : min of maximum
* ‘lom’ : max of maximum

## Operador de agregação (acumulação)
* accumulation_method: exemplos: np.fmax (padrão), np.multiply, np.min

```python
b = ctrl.Consequent(np.arange(y.min(), y.max(), 0.01), "B", defuzzify_method='centroid')
b['b1'] = fuzz.gaussmf(b.universe, 0, 0.125)
b['b2'] = fuzz.gaussmf(b.universe, 0.3125, 0.125)
b['b3'] = fuzz.gaussmf(b.universe, 0.625, 0.1875)
b['b4'] = fuzz.gaussmf(b.universe, 1.89, 0.44)
b['b5'] = fuzz.gaussmf(b.universe, 4, 0.609375)

b.view()
```

# Regras

## Tipos de operadores de AND e OR
* and_func : function, optional
            Function which accepts multiple floating-point arguments and
            returns a single value. Defalts to NumPy function `fmin`, to
            support both single values and arrays. For multiplication,
            substitute `fuzz.control.mult` or `np.multiply`.
* or_func : function, optional
    Function which accepts multiple floating-point arguments and
    returns a single value. Defalts to NumPy function `fmax`, to
    support both single values and arrays.

```python
and_func = np.multiply
or_func = np.fmax

r1 = ctrl.Rule(antecedent=a['a1'], consequent=b['b5'], label='R1', and_func=and_func, or_func=or_func)
r2 = ctrl.Rule(antecedent=a['a2'], consequent=b['b4'], label='R2', and_func=and_func, or_func=or_func)
r3 = ctrl.Rule(antecedent=a['a3'], consequent=b['b3'], label='R3', and_func=and_func, or_func=or_func)
r4 = ctrl.Rule(antecedent=a['a4'], consequent=b['b2'], label='R4', and_func=and_func, or_func=or_func)
r5 = ctrl.Rule(antecedent=a['a5'], consequent=b['b1'], label='R5', and_func=and_func, or_func=or_func)
r6 = ctrl.Rule(antecedent=a['a6'], consequent=b['b2'], label='R6', and_func=and_func, or_func=or_func)
r7 = ctrl.Rule(antecedent=a['a7'], consequent=b['b3'], label='R7', and_func=and_func, or_func=or_func)
r8 = ctrl.Rule(antecedent=a['a8'], consequent=b['b4'], label='R8', and_func=and_func, or_func=or_func)
r9 = ctrl.Rule(antecedent=a['a9'], consequent=b['b5'], label='R9', and_func=and_func, or_func=or_func)

control_system = ctrl.ControlSystem(rules = [r1, r2, r3, r4, r5, r6, r7, r8, r9])
simulation = ctrl.ControlSystemSimulation(control_system)

def simulate(simulation, xi):
    simulation.input['A'] = xi
    simulation.compute()
    print('z = ',simulation.output['B'])
    a.view(sim=simulation)
    b.view(sim=simulation)
```
